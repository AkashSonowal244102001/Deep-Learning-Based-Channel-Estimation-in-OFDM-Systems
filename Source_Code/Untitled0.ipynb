```python
from scipy.io import loadmat

def inspect_mat_file(path):
    """
    Load a .mat file and print each user variable's name, type, shape, and dtype.

    Parameters:
    -----------
    path : str
        Filesystem path to the .mat file.
    """
    mat = loadmat(path)
    print(f"Contents of '{path}':\n")
    for name, val in mat.items():
        # skip MATLAB metadata entries
        if name.startswith('__'):
            continue
        # discover shape and dtype if they exist
        shape = getattr(val, 'shape', None)
        dtype = getattr(val, 'dtype', None)
        print(f"  â€¢ {name!r}: type={type(val).__name__}, shape={shape}, dtype={dtype}")

# Example usage:
inspect_mat_file('Noisy_Veha_SNR_22.mat')
```
